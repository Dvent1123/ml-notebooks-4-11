# Categorical Correlations

```python
from scipy.stats import chi2_contingency
#using chi squared to analyze independence of categorical columns
#the further away from zero the less dependent the variable with respect to the target variable

contingency_table = data.groupby(['Warehouse_block','Reached.on.Time_Y.N']).size().reset_index() \
    .rename(columns={0:'Counts'}).pivot(columns='Warehouse_block',index='Reached.on.Time_Y.N')
print(contingency_table)

stat, p, dof, expected = chi2_contingency(contingency_table)
print(f'\nP value for independence: {p:.4f}')
```

```python
#chi squared to analyze independence of mode of shipment
#p value is far from zero so reached on time is not very dependent on mode of shipment

contigency_table = data.groupby(['Mode_of_Shipment', 'Reached.on.Time_Y.N']).size().reset_index() \
    .rename(columns={0: 'Counts'}).pivot(columns='Mode_of_Shipment',index='Reached.on.Time_Y.N')
print(contigency_table)
stat, p, dof, expected = chi2_contingency(contigency_table)
print(f'\nP value for independence: {p: .4f}')
```

# Preprocessing

```python
#function to pipeline encoding process
def onehot_encode(df, column):
    #copies data frame given into another data fram
    df=df.copy()
    #encodes column with prefix being column name
    dummies = pd.get_dummies(df[column], prefix=column)
    #concats the original data frame with the encoded columns returned in previous line
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    
    return df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#function for all preprocessing
def preprocess_inputs(df):
    df = df.copy()
    
    #drops the ID column since unneccessary
    
    df = df.drop('ID', axis=1)
    
    #binary encoding the gender column
    df['Gender'] = df['Gender'].replace({'F': 0, 'M': 1})
    
    #one-hot encoding using onehot_encode() in for loop
    for column in ['Warehouse_block', 'Mode_of_Shipment', 'Product_importance']:
        
        df = onehot_encode(df, column=column)
        
    #splits the data into X and y
    y = df['Reached.on.Time_Y.N']
    X = df.drop('Reached.on.Time_Y.N', axis=1)
    
    #split the test and training data
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
    
    #scale X using Standard Scalar
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = pd.DataFrame(scaler.transform(X_train), index=X_train.index,
                          columns=X_train.columns)
    #do same thing to testing set
    X_test = pd.DataFrame(scaler.transform(X_test), index=X_test.index,
                         columns=X_test.columns)
    
    return X_train, X_test, y_train, y_test
    
```

```python
df = pd.read_csv('./Train.csv')
data = df.copy()

X_train, X_test, y_train, y_test = preprocess_inputs(data)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
# from xgboost import XGBClassifier
# from lightgbm import LGBMClassifier
# from catboost import CatBoostClassifier

models = {
    "                   Logistic Regression": LogisticRegression(),
    "                   K-Nearest Neighbors": KNeighborsClassifier(),
    "                         Decision Tree": DecisionTreeClassifier(),
    "Support Vector Machine (Linear Kernel)": LinearSVC(),
    "   Support Vector Machine (RBF Kernel)": SVC(),
    "                        Neural Network": MLPClassifier(),
    "                         Random Forest": RandomForestClassifier(),
    "                     Gradient Boosting": GradientBoostingClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    print(name + ' trained.')
```

```python
for name, model in models.items():
    print(name + ': {:.2f}%'.format(model.score(X_test, y_test) * 100))
```

# Notes on What I Understand of different Models

## Logistic Regression

Logistic regression predicts true or false with an S-shaped curve and is used for classification. If you give it a new data point it predicts what the data should be classified as based on where it lands on the curve.

## K-nearest neighbors

K-nearest neighbor is good for scattered data and is another useful classification algorithm. It predicts where a new data point will be clusterd into (or classified as). The algorithm looks at a k amount of neighbors that we assign it and looks at what those "neighbors" are classified under and then decides where the new data point should be classified as based on the data

## Decision Tree

Decision trees are another classifier. It starts at the root node and then uses numerical as well as true or false questions to get to the bottom. The bottom is what the new data point gets classified as. If the root node has multiple options you can calculate which is the best by calculating Gini impurity and the one with the smallest value is the one that classifies data points the best and should thus be the root node for the tree.
